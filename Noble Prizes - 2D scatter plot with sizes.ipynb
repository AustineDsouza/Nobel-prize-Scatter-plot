```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load noble prize dataset into pandas dataframe
noble = pd.read_excel('C:/Austine/Blog/ML/Reference/Noble_prize.xlsx')
```

```python
#Print the number of data point and features
print(noble.shape)
```

```python
#What are the column names in our dataset?
print (noble.columns)
```

```python
#Simple 2D scatter plots without color coding
#Always understand the axis: labels and scale

noble.plot(kind='scatter', x='Country_index', y='No.of Noble prizes won');
plt.show()
```

```python
#Scatter plot with color coding
#Here 'sns' corresponds to seaborn

sns.set_style("whitegrid");
sns.FacetGrid(noble, hue = "Country name", size = 7) \
    .map(plt.scatter, "Country_index", "No.of Noble prizes won") \
    .add_legend();
plt.show();
```

```python
#We can go further and also add size to our plot
#The plot will now show the datapoints with the size of the defined column
```

```python
noble = np.array(noble)
new = np.copy(noble)
new = new.T
```

```python
size = new[3]
size = 0.00001*size.astype(int)

ax = plt.figure(figsize=(15,8.5))
plt.scatter(new[1], new[2], alpha=0.2, s=size,
             c = new[3], cmap='viridis')
plt.xlabel('COUNTRY INDEX')
plt.ylabel('NOI. OF NOBLE PRIZES WON');

for i, txt in enumerate(new[0]):
    plt.annotate(txt, (new[1][i],new[2][i]))

plt.colorbar()
plt.show();
```

```python
#Let's drop the outperforming country United States and see how our data looks like
noble1 = np.array(noble)
noble1 = noble[0:29]
new1 = np.copy(noble1)
new1 = new1.T
```

```python
size = new1[3]
size = 0.0001*size.astype(int)

ax = plt.figure(figsize=(15,8.5))
plt.scatter(new1[1], new1[2], alpha=0.2, s=size,
             c = new1[3], cmap='viridis')
plt.xlabel('COUNTRY INDEX')
plt.ylabel('NOI. OF NOBLE PRIZES WON');

for i, txt in enumerate(new1[0]):
    plt.annotate(txt, (new1[1][i],new1[2][i]))

plt.colorbar()
plt.show();
```
